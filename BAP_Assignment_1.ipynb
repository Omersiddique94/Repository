#### 7.  Iteration and Conditioning
Write an iterative program (for loop) to determine the maximum value in a list. Do not use the max() function.

````
def check_max(lst):
    # your code
    x=0
    for item in lst:
      if item>x:
        x = item
    print (x)


# define some sample values
nums = [3, 41, 12, 9, 74, 15]

# run the program
check_max(nums)
````
